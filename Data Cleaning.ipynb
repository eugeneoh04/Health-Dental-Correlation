Rename Columns

```python
col_dict = ['HCHK_YEAR', 'IDV_ID', 'SIDO', 'SEX', 'AGE_GROUP', 
            'HEIGHT', 'WEIGHT', 'WAIST', 'SIGHT_LEFT', 'SIGHT_RIGHT',
            'HEAR_LEFT', 'HEAR_RIGHT', 'BP_HIGH', 'BP_LWST',
            'BLDS', 'TOT_CHOLE', 'TRIGLYCERIDE', 'HDL_CHOLE', 'LDL_CHOLE',
            'HMG', 'OLIG_PROTE_CD', 'CREATININE', 'SGOT_AST',
            'SGPT_ALT', 'GAMMA_GTP', 'SMK_STAT_TYPE_CD', 'DRK_YN',
            'HCHK_OE_INSPEC_YN', 'CRS_YN', 'TTR_YN', 'DATA_STD__DT']
```

```python
df_rn = df.copy()
df_rn.columns = col_dict
```

Clean Data

```python
df_cl = df_rn.copy()
```

```python
#validate all values are from a unique patient
df_cl['IDV_ID'].nunique()
```

```python
#drop irrelevant columns
df_cl = df_cl.drop(columns=['HCHK_YEAR', 'IDV_ID', 'HEAR_LEFT', 'HEAR_RIGHT', 'SIDO', 'HEIGHT', 'WEIGHT', 'WAIST', 'SIGHT_LEFT', 'SIGHT_RIGHT', 'TTR_YN', 'DATA_STD__DT'])
df_cl.columns
```

```python
df_cl = df_cl.dropna()
df_cl.shape[0]
```
